```python
import spacy

# Load the transformer-based English model
nlp = spacy.load("en_core_web_trf")
```

```python
file_path = "/finiteCCs_cleaned_final.txt"
```

```python
with open(file_path, "r", encoding="utf-8") as file:
  finite_ccs = file.read()
```

```python
doc = nlp(finite_ccs)
for sent in doc.sents:
  print(sent)
```

```python
def contains_negation(doc):
    """
    Check if a sentence contains a negation marker.
    Returns 1 if negation is present, 0 otherwise.
    """
    negation_markers = {"not", "no", "never", "none", "nothing", "nobody", "nowhere", "neither", "nor"}

    for token in doc:
        # Check if the token is a negation marker
        if token.text.lower() in negation_markers:
            return 1
        # Check for dependency-based negation (e.g., "doesn't", "isn't")
        if token.dep_ == "neg":
            return 1
    return 0
```

```python
def process_sentences(input_file, output_file):
    """
    Process sentences from the input file and determine the presence or absence of negation markers.
    Write the results to the output file with two columns: sentence and negation flag (0 or 1).
    """
    with open(input_file, "r", encoding="utf-8") as infile, \
         open(output_file, "w", encoding="utf-8") as outfile:

        # Write header to the output file
        outfile.write("Clause\tNegation Flag\n")

        # Process each line in the input file
        for line in infile:
            # Strip leading/trailing whitespace
            sentence = line.strip()

            # Skip empty lines
            if not sentence:
                continue

            # Parse the sentence with spaCy
            doc = nlp(sentence)

            # Check for negation markers
            negation_flag = contains_negation(doc)

            # Write the result to the output file
            outfile.write(f"{sentence}\t{negation_flag}\n")

            # Print the result to the console
            print(f"Clause: {sentence}")
            print(f"Negation Flag: {negation_flag}\n")

    print(f"Results have been written to {output_file}")

# That-Clauses
# input_file = "/finiteCCs_cleaned_final.txt"  # Input file path
# output_file = "/negation_that-cl.tsv"  # Output file path

# To-Inf
input_file = "/all-nonfinCC-very-final-nodup.txt"
output_file = "/negation_to-inf.tsv"
process_sentences(input_file, output_file)
```
