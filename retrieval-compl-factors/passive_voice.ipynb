```python
import spacy

# Load the transformer-based English model
nlp = spacy.load("en_core_web_trf")
```

```python
def is_passive_voice(doc):
    """
    Check if a sentence is in passive voice.
    Passive voice is identified by:
    1. A form of the verb "be" (e.g., "is", "was", "were", "has been").
    2. The next token after "be" is a past participle verb form (e.g., "eaten", "written", "taken").
    """
    for i, token in enumerate(doc):
        # Check if the token is a form of "to be"
        if token.lemma_ == "be":
            # Check if the next token exists and is a past participle
            if i + 1 < len(doc) and doc[i + 1].tag_ == "VBN":
                return 1
    return 0
```

```python
def process_sentences(input_file, output_file):
    """
    Process sentences from the input file and determine the presence or absence of passivization.
    Write the results to the output file with two columns: sentence and passivization (0 or 1).
    """
    with open(input_file, "r", encoding="utf-8") as infile, \
         open(output_file, "w", encoding="utf-8") as outfile:

        # Write header to the output file
        outfile.write("Clause\tPassivization\n")

        # Process each line in the input file
        for line in infile:
            # Strip leading/trailing whitespace
            sentence = line.strip()

            # Skip empty lines
            if not sentence:
                continue

            # Parse the sentence with spaCy
            doc = nlp(sentence)

            passive_flag = is_passive_voice(doc)

            # Write the result to the output file
            outfile.write(f"{sentence}\t{passive_flag}\n")

            # Print the result to the console
            print(f"Clause: {sentence}")
            print(f"Passive Voice Flag: {passive_flag}\n")

    print(f"Results have been written to {output_file}")

# That-Clauses
input_file = "/finiteCCs_cleaned_final.txt"  # Input file path
output_file = "/passive_that-cl.tsv"  # Output file path

# To-Inf
# input_file = "/all-nonfinCC-very-final-nodup.txt"
# output_file = "/passive_to-inf.tsv"
process_sentences(input_file, output_file)
```
