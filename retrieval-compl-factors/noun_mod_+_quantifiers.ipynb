```python
import spacy

# Load the transformer-based English model
nlp = spacy.load("en_core_web_trf")
```

```python
def contains_nom_modifiers(doc):

    for token in doc:
        # Check if the
        if token.dep_ in {"appos", "nmod", "amod"}:
            return 1
    return 0
```

```python
def contains_quantifiers(doc):

    for token in doc:
      if token.dep_ == "nummod":
        return 1
      if token.pos_ == "DET" and token.text.lower() in {"some", "many", "much", "several", "a few", "few", "both", "all", "a little", "little", "more", "less", "enough"}:
        return 1
      if token.pos_ != "DET" and token.text.lower() in {"some", "many", "much", "several", "a few", "few", "both", "all", "all of the", "a little", "little", "number of", " majority of", "third of", "quarter of", "none", "none of the"}:
        return 1
    return 0
```

```python
def process_sentences(input_file, output_file):


    with open(input_file, "r", encoding="utf-8") as infile, \
         open(output_file, "w", encoding="utf-8") as outfile:

        # Write header to the output file
        outfile.write("Clause\tNoun-Mod\tQuantifier\n")

        # Process each line in the input file
        for line in infile:
            # Strip leading/trailing whitespace
            sentence = line.strip()

            # Skip empty lines
            if not sentence:
                continue

            # Parse the sentence with spaCy
            doc = nlp(sentence)

            # Check for noun modifiers and quantifiers
            nom_mod_flag = contains_nom_modifiers(doc)
            quant_flag = contains_quantifiers(doc)
            # Write the result to the output file
            outfile.write(f"{sentence}\t{nom_mod_flag}\t{quant_flag}\n")

            # Print the result to the console
            print(f"Clause: {sentence}")
            print(f"Nom-Mod Flag: {nom_mod_flag}\n")
            print(f"Quantifier Flag: {quant_flag}\n")

    print(f"Results have been written to {output_file}")

# That-Clauses
# input_file = "/finiteCCs_cleaned_final.txt"
# output_file = "/nom-mod-quant_that-cl.tsv"

# To-Inf
input_file = "/all-nonfinCC-very-final-no-infm.txt"
output_file = "/nom_mod + quantifiers_to-inf.tsv"
process_sentences(input_file, output_file)
```
